```python
import os
import pandas as pd

ROOT = r"C:\Users\lgilb\fantasyfootball\sleeper_dynasty_adp\scripts_or_notebooks\sleeper_dynasty_adp"
SNAP = os.path.join(ROOT, "data", "snapshots")

draft_catalog_path = os.path.join(SNAP, "draft_catalog", "draft_catalog_ALL.parquet")
dc = pd.read_parquet(draft_catalog_path)

# If your catalog has a season column, filter it; otherwise just write ALL to the chosen season folder
SEASON = 2026

# best effort filtering
if "season" in dc.columns:
    drafts = dc[dc["season"].astype(str) == str(SEASON)].copy()
elif "season_year" in dc.columns:
    drafts = dc[dc["season_year"].astype(str) == str(SEASON)].copy()
else:
    drafts = dc.copy()

out_folder = os.path.join(SNAP, "drafts", f"season={SEASON}")
os.makedirs(out_folder, exist_ok=True)

out_path = os.path.join(out_folder, "drafts.parquet")
drafts.to_parquet(out_path, index=False)

print("[OK] wrote drafts snapshot:", out_path, "| shape:", drafts.shape)
print("cols:", list(drafts.columns)[:40])
```

```python
import os
import glob

ROOT = r"C:\Users\lgilb\fantasyfootball\sleeper_dynasty_adp\scripts_or_notebooks\sleeper_dynasty_adp"
RAW  = os.path.join(ROOT, "data", "raw")
SNAP = os.path.join(ROOT, "data", "snapshots")

def show(patterns, base, limit=50):
    hits = []
    for pat in patterns:
        hits += glob.glob(os.path.join(base, "**", pat), recursive=True)
    hits = sorted(set(hits))
    print("\nBASE:", base)
    print("FOUND:", len(hits))
    for p in hits[:limit]:
        print(" -", p)
    if len(hits) > limit:
        print(f"... (+{len(hits)-limit} more)")
    return hits

raw_hits = show(
    patterns=["*picks*.parquet", "*picks*.csv", "*leagues*.parquet", "*leagues*.csv", "*drafts*.parquet", "*drafts*.csv"],
    base=RAW
)

snap_hits = show(
    patterns=["picks.parquet", "leagues.parquet"],
    base=SNAP
)
```
